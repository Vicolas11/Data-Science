```python
A = raindf[raindf['SUBDIVISION'].str.contains('ANDAMAN & NICOBAR ISLANDS')]
B = raindf[raindf.SUBDIVISION == 'BIHAR']
G = raindf[raindf['SUBDIVISION'].str.contains('GUJARAT REGION')]
K = raindf[raindf.SUBDIVISION == 'KERALA']
W = raindf[raindf['SUBDIVISION'].str.contains('WEST RAJASTHAN')]
```

```python
plt.figure(figsize=(15,15))
plt.plot(A['YEAR'], A['ANNUAL'], linestyle='--', marker='x', color='r', label='Andaman')
plt.plot(B['YEAR'], B['ANNUAL'], linestyle='--', marker='x', color='b', label='Bihar')
plt.plot(G['YEAR'], G['ANNUAL'], linestyle='--', marker='x', color='g', label='Gujarat')
plt.plot(K['YEAR'], K['ANNUAL'], linestyle='--', marker='x', color='y', label='Kerala')
plt.plot(W['YEAR'], W['ANNUAL'], linestyle='--', marker='x', color='black', label='W Rajasthan')

plt.xlabel('Year')
plt.ylabel('Annual RainFall')
plt.title('India Annual Rainfall Time Series')
plt.legend()
plt.show()
```
